```python
import cv2
from ultralytics import YOLO

model = YOLO("best.pt")  

cap = cv2.VideoCapture(0) 

while cap.isOpened():
    ret, frame = cap.read() 
    if not ret:
        break

    results = model.predict(source=frame, show=True)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
